## 1) Plot the function $y = 3x^2$ for $1 ≤ x ≤ 3$ as a continuous line. Include enough points so that the curve you plot appears smooth. Label the axes $x$ and $y$.

## 2) Plot the following function for $−15 ≤ x ≤ 15$:


$y = \frac{cosx}{1 + \frac{1}{5}x^2}$

## Include enough points so that the curve you plot appears smooth. Draw thin gray lines, one horizontal at $y = 0$ and the other vertical at $x = 0$. Both lines should appear behind the function. Label the axes $x$ and $y$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-8*np.pi, 8*np.pi, 1000)
y = np.cos(x)/(1 + (1/5)*(x**2))

plt.figure(1, figsize=(10, 6))
plt.axhline(color='gray', linestyle='dashed', zorder=-1)
plt.axvline(color='gray', zorder=-1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.plot(x,y)
plt.show()
```

## 5) Use $matplotlib's$ function $hist$ along with Numpy's functions $random.rand$ and $random.randn$ to create a histogram graphs in Figure 9.2.  See §9.2 for a description of NumPy’s random number functions.

## b) In a single window frame, make three vertically stacked plots of the displacement, velocity, and acceleration vs. time. Show the error bars on the different plots. Make sure that the time axes of all three plots cover the same range of times (use sharex). Why do the relative sizes of the error bars grow progressively greater as one progresses from displacement to velocity to acceleration?




```python
import numpy as np
import matplotlib.pyplot as plt

t, d, dd = np.loadtxt('data_exer7.txt', skiprows=4, unpack=True)

v = (d[1:] - d[:-1])/(t[1:] - t[:-1])
tv = (t[1:] + t[:-1])/2


a = (v[1:] - v[:-1])/(tv[1:] - tv[:-1])
ta = (tv[1:] + tv[:-1])/2

dv = np.sqrt(((1/(t[1:]-t[:-1]))*dd[1:])**2+((1/(t[1:]-t[:-1]))*dd[:-1])**2)
da = np.sqrt(((1/(tv[1:]-tv[:-1]))*dv[1:])**2+((1/(tv[1:]-tv[:-1]))*dv[:-1])**2)



fig = plt.figure(figsize=(9, 7))

ax1 = fig.add_subplot(3,1,1)
ax1.plot(t, d, 'g--')
ax1.errorbar(t, d, fmt='oC1', label='data', yerr = dd, ecolor='black' )
ax1.set_xlabel('Time(s)')
ax1.set_ylabel('Displacement(m)')

ax2 = fig.add_subplot(3, 1, 2)
ax2.plot(tv, v, 'b--')
ax2.errorbar(tv, v, fmt='oC2', label='data', yerr = dv, ecolor='black')
ax2.set_xlabel('Time(s)')
ax2.set_ylabel('Velocity(m/s)')

ax3 = fig.add_subplot(3, 1, 3)
ax3.plot(ta, a, 'r--')
ax3.errorbar(ta, a, fmt='oC2', label='data', yerr = da, ecolor='black')
ax3.set_xlabel('Time(s)')
ax3.set_ylabel('Acceleration (m/s2)')


fig.tight_layout()
fig.show()



#Uncertainties grow
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.arange(0.01, 8., 0.04)
y = np.sqrt((8./theta)**2 - 1)
ytan = np.tan(theta)
ytan = np.ma.masked_where(np.abs(ytan) > 20., ytan)

ycot = 1./np.tan(theta)
ycot = np.ma.masked_where(np.abs(ycot) > 20., ycot)

plt.close()


plt.rc('mathtext', fontset='stix')

fig, ax = plt.subplots(2, 1, figsize=(8.5, 6), sharex=True, sharey=True)

for i in range(2):
    
    if i == 0:
        ax[i].plot(theta, y, linestyle=':')
        ax[i].plot(theta, ytan)
        
    else:
        ax[i].plot(theta, -y, linestyle=':')
        ax[i].plot(theta, ycot)
        
    ax[i].set_xlim(0, 8)
    ax[i].set_ylim(-8, 8)
    ax[i].axhline(color='gray', zorder=-1)
    ax[i].axvline(x=np.pi/2., color='gray', linestyle='--', zorder=-1)
    ax[i].axvline(x=3*np.pi/2., color='gray', linestyle='--', zorder=-1)
    ax[i].axvline(x=5*np.pi/2., color='gray', linestyle='--', zorder=-1)
    ax[i].set_xlabel(r'$\theta$')
    ax[i].set_ylabel(r'$Tan\theta$')
    
plt.savefig('exercise6_7')








```
