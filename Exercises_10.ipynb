Your code should find the average $a$, maximum $m$, and standard deviation $s$ of the value of each currency relative to the US dollar over the period of time starting from the first business day of 2017, January $3^{rd}$. The DataFrame should be sorted from the largest $mx$ to the smallest.

```python
import pandas as pd

url1 = 'http://www.bankofcanada.ca/'
url2 = 'valet/observations/group/FX_RATES_DAILY/csv?start_date='
fecha = '2017-01-03'

url = url1 + url2 + fecha

 
data = pd.read_csv(url, skiprows=39, index_col= 'date')
days, currencies = data.shape
data
```

```python
codes = pd.read_csv(url, skiprows=10, usecols=[0,2], 
                   nrows=currencies)
codes
```

```python

#First we must calculate the average a and the maximum mx of each currency relative to the CAD



#Let's create a list for the code names
ids = []

for c in codes['id']:
    ids.append(c)

print(ids)

#Let's create a list for the variables mx

lst_mx = []

for (col, cod) in zip(range(data.shape[1]), ids):
    
    promedio = sum(data[cod])/(data.shape[0])
    mx = max(data[cod])/promedio
    
    lst_mx.append(mx)

print(lst_mx)

```
