## 1) Plot the function $y = 3x^2$ for $1 ≤ x ≤ 3$ as a continuous line. Include enough points so that the curve you plot appears smooth. Label the axes $x$ and $y$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1., 3., 200)
y = 3*(x**2)

plt.title('Exercise 6')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.show()
```

## 2) Plot the following function for $−15 ≤ x ≤ 15$:


$y = \frac{cosx}{1 + \frac{1}{5}x^2}$

## Include enough points so that the curve you plot appears smooth. Draw thin gray lines, one horizontal at $y = 0$ and the other vertical at $x = 0$. Both lines should appear behind the function. Label the axes $x$ and $y$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-8*np.pi, 8*np.pi, 1000)
y = np.cos(x)/(1 + (1/5)*(x**2))

plt.figure(1, figsize=(10, 6))
plt.axhline(color='gray', linestyle='dashed', zorder=-1)
plt.axvline(color='gray', zorder=-1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.plot(x,y)
plt.show()
```

## 3) Plot the functions $sin(x)$ and $cos(x)$ vs $x$ on the same plot with $x$ going from $-\pi$ to $\pi$. Make sure the limits of the x-axis do not extend beyond the limits of the data. Plot $sin(x)$ in the color orange and $cos(x)$ in the color green and include a legend to label the two curves. Place the legend within the plot, but such that it does not cover either of the sine or cosine traces. Draw thin gray lines behind the curves, one horizontal at $y = 0$ and the other vertical at $x = 0$.


```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-4*np.pi, 4*np.pi, 1000)
f1 = np.cos(x)
f2 = np.sin(x)
#f3 = np.tan(x)

#f3 = np.ma.masked_where(np.abs(f3) > 4, f3)

plt.figure(figsize=(8, 6))
plt.axhline(color='gray', linestyle='--', zorder=-1)
plt.axvline(color='gray', linestyle='--', zorder=-1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.xlim(-4*np.pi, 4*np.pi)
#plt.plot(x, f3, label='Tan')
plt.plot(x, f2, color='orange', label='Cos')
plt.plot(x, f1, color='green', label='Sen')
plt.legend(loc='upper right', title='Gráficas')
```

### a) Read the data into the Python program and plot $t$ vs $y$ using circles for data points with error bars. Use the data in the $dy$ column as the error estimates for the $y$ data. Label the horizontal and vertical axes $“time (s)”$ and $“position (cm)”$. Create your plot using the $fig, ax = plt.subplots()$.

```python
import numpy as np
import matplotlib.pyplot as plt

time, position, error_y = np.loadtxt('data_exer6.txt', skiprows=5, unpack=True)

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlabel('time(s)')
ax.set_ylabel('Position(cm)')
ax.errorbar(time, position, fmt='oC0', label='data', yerr=error_y, ecolor='black')
```

## b) On the same graph, plot the function below as a smooth line. Make the line pass behind the data points.

$y(t) = \left(3 + \frac{1}{2}sin\frac{\pi t}{5}\right)te^{-t/10}$

```python
import numpy as np
import matplotlib.pyplot as plt

time, position, error_y = np.loadtxt('data_exer6.txt', skiprows=5, unpack=True)

x = np.linspace(0, 50, 100)
y = (3 + 0.5*np.sin(np.pi*x/5))*x*np.exp(-x/10)


fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, color='red', label='Model')
ax.set_xlabel('time(s)')
ax.set_ylabel('Position(cm)')
ax.errorbar(time, position, fmt='oC0', label='Data', yerr=error_y, ecolor='black')
ax.legend(loc='upper right')




```
