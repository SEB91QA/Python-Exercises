## 1) Plot the function $y = 3x^2$ for $1 ≤ x ≤ 3$ as a continuous line. Include enough points so that the curve you plot appears smooth. Label the axes $x$ and $y$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1., 3., 200)
y = 3*(x**2)

plt.title('Exercise 6')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.show()
```

## 2) Plot the following function for $−15 ≤ x ≤ 15$:


$y = \frac{cosx}{1 + \frac{1}{5}x^2}$

## Include enough points so that the curve you plot appears smooth. Draw thin gray lines, one horizontal at $y = 0$ and the other vertical at $x = 0$. Both lines should appear behind the function. Label the axes $x$ and $y$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-8*np.pi, 8*np.pi, 1000)
y = np.cos(x)/(1 + (1/5)*(x**2))

plt.figure(1, figsize=(10, 6))
plt.axhline(color='gray', linestyle='dashed', zorder=-1)
plt.axvline(color='gray', zorder=-1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.plot(x,y)
plt.show()
```

## 3) Plot the functions $sin(x)$ and $cos(x)$ vs $x$ on the same plot with $x$ going from $-\pi$ to $\pi$. Make sure the limits of the x-axis do not extend beyond the limits of the data. Plot $sin(x)$ in the color orange and $cos(x)$ in the color green and include a legend to label the two curves. Place the legend within the plot, but such that it does not cover either of the sine or cosine traces. Draw thin gray lines behind the curves, one horizontal at $y = 0$ and the other vertical at $x = 0$.


```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-4*np.pi, 4*np.pi, 1000)
f1 = np.cos(x)
f2 = np.sin(x)
#f3 = np.tan(x)

#f3 = np.ma.masked_where(np.abs(f3) > 4, f3)

plt.figure(figsize=(8, 6))
plt.axhline(color='gray', linestyle='--', zorder=-1)
plt.axvline(color='gray', linestyle='--', zorder=-1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.xlim(-4*np.pi, 4*np.pi)
#plt.plot(x, f3, label='Tan')
plt.plot(x, f2, color='orange', label='Cos')
plt.plot(x, f1, color='green', label='Sen')
plt.legend(loc='upper right', title='Gráficas')
```

## b) On the same graph, plot the function below as a smooth line. Make the line pass behind the data points.

$y(t) = \left(3 + \frac{1}{2}sin\frac{\pi t}{5}\right)te^{-t/10}$

```python
import numpy as np
import matplotlib.pyplot as plt

time, position, error_y = np.loadtxt('data_exer6.txt', skiprows=5, unpack=True)

x = np.linspace(0, 50, 100)
y = (3 + 0.5*np.sin(np.pi*x/5))*x*np.exp(-x/10)


fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, color='red', label='Model')
ax.set_xlabel('time(s)')
ax.set_ylabel('Position(cm)')
ax.errorbar(time, position, fmt='oC0', label='Data', yerr=error_y, ecolor='black')
ax.legend(loc='upper right')




```

## 5) Use $matplotlib's$ function $hist$ along with Numpy's functions $random.rand$ and $random.randn$ to create a histogram graphs in Figure 9.2.  See §9.2 for a description of NumPy’s random number functions.

```python
import numpy as np
import matplotlib.pyplot as plt


fig, ax = plt.subplots(1, 2, figsize=(12.5, 5.5))

x  = np.random.rand(10000)
xn = np.random.randn(10000)

ax[0].hist(x, 100, facecolor='blue', alpha=0.5)
ax[0].text(0.45, 0.96, 
       'np.rand(10000)',
       fontsize=16, ha='right', va='top',
       transform=ax[0].transAxes)
ax[1].hist(xn, 100, facecolor='blue', alpha=0.5)
ax[1].text(0.45, 0.96,
           'np.randn(10000)',
           fontsize=14, ha='right', va='top',
           transform=ax[1].transAxes)
plt.xlim(-4, 4)
plt.ylim(0, 350)
plt.show()

```

## b) In a single window frame, make three vertically stacked plots of the displacement, velocity, and acceleration vs. time. Show the error bars on the different plots. Make sure that the time axes of all three plots cover the same range of times (use sharex). Why do the relative sizes of the error bars grow progressively greater as one progresses from displacement to velocity to acceleration?




```python
import numpy as np
import matplotlib.pyplot as plt

t, d, dd = np.loadtxt('data_exer7.txt', skiprows=4, unpack=True)

v = (d[1:] - d[:-1])/(t[1:] - t[:-1])
tv = (t[1:] + t[:-1])/2


a = (v[1:] - v[:-1])/(tv[1:] - tv[:-1])
ta = (tv[1:] + tv[:-1])/2

dv = np.sqrt(((1/(t[1:]-t[:-1]))*dd[1:])**2+((1/(t[1:]-t[:-1]))*dd[:-1])**2)



plt.rc('mathtext', fontset='stix') 

fig = plt.figure(figsize=(9, 7))

ax1 = fig.add_subplot(3,1,1)
ax1.plot(t, d, 'g--')
ax1.errorbar(t, d, fmt='oC1', label='data', yerr = dd, ecolor='black' )
ax1.set_xlabel('Time(s)')
ax1.set_ylabel('Displacement(m)')

ax2 = fig.add_subplot(3, 1, 2)
ax2.plot(tv, v, 'b--')
ax2.errorbar(tv, v, fmt='oC2', label='data', yerr = dv, ecolor='black')
ax2.set_xlabel('Time(s)')
ax2.set_ylabel('Velocity(m/s)')

ax3 = fig.add_subplot(3, 1, 3)
ax3.plot(ta, a, 'r--')
ax3.errorbar(ta, a, fmt='oC2', label='data', yerr = da, ecolor='black')
ax3.set_xlabel('Time(s)')
ax3.set_ylabel('Acceleration (m/s2)')


fig.tight_layout()
fig.show()
```
