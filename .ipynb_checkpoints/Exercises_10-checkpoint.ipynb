## 1. Read the planetary data in the text file $planetData.txt$ into a Pandas DataFrame and perform the following tasks.

## (a) Based on the data read in from the file $planetData.txt$, find the average density of each planet relative to that of the Earth and add the results as a column in your DataFrame.

## (b) Print out your DataFrame sorted from the largest to smallest diameter planet.

## (c) Make a list of planets that have masses greater than that of Earth, sorted from least to most massive planet.

```python
#(a)

import pandas as pd
import numpy as np

#Read in the data file
planets = pd.read_table('planetData.txt', sep='\s+')
planets
```

```python
#Let's add the colummn

Volume = (4./3.)*np.pi*(planets['diameter']/2.)**3

planets['Density'] = planets['mass']/Volume
planets

#Density of the earth
Volumearh = (4./3.)*np.pi*(1.00/2.)**3

planets['DensityR'] = planets['Density']/Volumearh
planets
```

```python
#(b)
planets.sort_values(by='diameter', ascending=False)
```

```python
#(c)

mass = planets.loc[planets['mass'] > planets['mass']['Earth'], 'mass']
pd.DataFrame(mass)
```
